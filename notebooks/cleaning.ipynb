```python
df_cartoes = pd.DataFrame(data=cartoes)
df_cartoes.head()
```

```python
df_cartoes["cartao"] = df_cartoes.values.tolist()
df_cartoes.head()
```

```python
df_cartoes[['11ac', '12ac','13ac','14ac','15ac']] = 0
df_cartoes
```

```python
acertos_all = []
for idx_cart, row_cart in tqdm(df_cartoes.iterrows()):
    acertos_cartao = []
    for idx, row in sorteados.iterrows():
        acertos_concurso = 0
        for numero in row_cart.cartao:
            acertos_concurso += row.sorteados_real.count(numero)
        if acertos_concurso > 10:
            col_name = str(acertos_concurso)+"ac"
            df_cartoes.loc[idx_cart, col_name] += 1
        # acertos_cartao.append(acertos_concurso)
    # for i in [11,12,13,14,15]:
    #     contagem_acertos.append(acertos_cartao.count(i))
    # acertos_all.append(contagem_acertos)
    
# print(*acertos_all, sep='|')
```

```python
df_cartoes.to_csv("../data/cartoes_20.csv")
```

```python
topCartoes15num = []
for index in list(df_cartoes[['cartao','11ac','12ac','13ac','14ac','15ac']].sort_values(by=['15ac','14ac','13ac','12ac','11ac'], ascending=False).head(10).index.values):
    for subset in itertools.combinations(df_cartoes.loc[index, 'cartao'], 15):
        topCartoes15num.append(subset)
len(topCartoes15num)
```

```python
acertos_all = []
for cartao_idx, cartao in enumerate(tqdm(topCartoes15num)):
    acertos_cartao = []
    contagem_acertos = []
    for idx, row in sorteados.iterrows():
        acertos_concurso = 0
        for numero in cartao:
            acertos_concurso += row.sorteados_real.count(numero)
        acertos_cartao.append(acertos_concurso)
    for i in [11,12,13,14,15]:
        contagem_acertos.append(acertos_cartao.count(i))
    acertos_all.append(contagem_acertos)
    
print(*acertos_all, sep='|')
```

```python
cartoesdf = pd.DataFrame(topCartoes15num)
cartoesdf['cartao'] = cartoesdf.values.tolist()
cartoesdf = cartoesdf['cartao']
acertos_df = pd.DataFrame(acertos_all, columns = ['11','12','13','14','15'])
acertos_df['cartao'] = cartoesdf
acertos_df
```

```python
acertos_df.to_csv('../data/topcartoes.csv')
```

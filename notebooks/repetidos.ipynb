## Geração do DF com probabilidades por número

```python
repeticoesDict = {}
for i in range(1,26):
    repeticoesDict[i] = {}
    repeticoesDict[i]['isolado'] = 0
    repeticoesDict[i][0] = 0
    for j in range(1,21):
        repeticoesDict[i][j] = 0
        col = f'x{j}repetidos'
        for idx, row in df.iterrows():
            print(f'Numero {i:02d} | Repeticoes {j:2d} | Linha {idx:4d}', end = '\r')
            if row[col] != '[]':
                if i in [int(x) for x in row[col].replace('[', '').replace(']', '').replace(' ','').split(',')]:
                    repeticoesDict[i][j] += 1
    for idx, row in df.iterrows():
        if i in row.sorteados_seq:
            repeticoesDict[i]['isolado'] += 1
            if row['x1repetidos'] != '[]':
                if 1 not in [int(x) for x in row['x1repetidos'].replace('[', '').replace(']', '').replace(' ','').split(',')]:
                    repeticoesDict[i][0] += 1

for i in range(1,26):
    acum = 0
    for j in range(20,1, -1):
        acum = acum + repeticoesDict[i][j]
        repeticoesDict[i][j-1] = (repeticoesDict[i][j-1] - acum) / len(df)

    repeticoesDict[i][0] = repeticoesDict[i][0] / repeticoesDict[i]['isolado']
    repeticoesDict[i]['isolado'] = repeticoesDict[i]['isolado'] / len(df) 


dfRep = pd.DataFrame(repeticoesDict)
dfRep
```

```python
def proximoCartao(df, index):
    sequenciasRepetidas = []
    for i in range(1,26):    
        sequencia = 0
        for j in range(index, index-21, -1):
            if i in df.loc[j, 'sorteados_seq']:
                sequencia += 1
            else:
                break
        sequenciasRepetidas.append([i,sequencia])

    numsProbs = []
    for i in sequenciasRepetidas:
        if i[1] > 0:
            divisor = probRepetir
        elif i[1] == 0:
            divisor = probNovos
        probabilidade = float(dfRep.loc[i[1], i[0]] * dfRep.loc['isolado', i[0]] / divisor)
        numsProbs.append([i[0], probabilidade])
    dfProbs = pd.DataFrame(numsProbs, columns = ['numero','probabilidade'])
    dfProbs.sort_values(by='probabilidade',ascending=False, inplace=True)
    dfProbs.reset_index(inplace=True,drop=True)
    dfProbs.sort_values(by='numero',ascending=True, inplace=True)
    proximoJogo = list(dfProbs[dfProbs.index<15]['numero'].values)
    return proximoJogo

def testaCartao(cartao, sorteio):
    acertos = 0
    for i in cartao:
        if i in sorteio:
            acertos+=1
    return acertos
```

## Teste do modelo

```python
resultados = []
for idx, row in tqdm(df[21:].iterrows()):
    acertos = testaCartao(proximoCartao(df, idx-1), row.sorteados_seq)
    col = f'Valor Rateio {acertos} Números'
    ganho = df.loc[idx, col] if acertos > 10 else 0
    custo = df.loc[idx, 'Valor Rateio 11 Números'] / 2
    resultados.append([row.Concurso, acertos, ganho, custo])
dfResultados = pd.DataFrame(resultados, columns=['Concurso', 'Acertos', 'Premio', 'Custo'])
```

## Gerando 100 cartões aleatórios e vendo o desempenho:

```python
import random

boasApostas = []

def mesmoCartaoAleatorio():
    cartao = random.sample(range(1,26), 15)
    ganho = 0
    gasto = 0
    custo = 0
    for idx, row in df.iterrows():
        acertos = 0
        for n in cartao:
            if n in row.sorteados_seq:
                acertos += 1
        col = f'Valor Rateio {acertos} Números'
        ganho += df.loc[idx, col] if acertos > 10 else 0
        custo += df.loc[idx, 'Valor Rateio 11 Números'] / 2
        if acertos == 15:
            boasApostas.append([row.Concurso, cartao, acertos, df.loc[idx, col]])
    saida = [cartao, ganho, custo, ((ganho/custo)-1)]  
    return saida

cartoesAleatorios = []
for i in tqdm(range(0,1000)):
    cartoesAleatorios.append(mesmoCartaoAleatorio())
    
dfCartoesAleatorios = pd.DataFrame(cartoesAleatorios, columns=['cartao', 'ganho', 'gasto', 'rendimento'])
```

```python
print(f'A mediana dos 100 cartões foi {np.median(dfCartoesAleatorios["rendimento"]):2.2%}')
print(f'A máxima dos 100 cartões foi {np.max(dfCartoesAleatorios["rendimento"]):2.2%}')
print(f'A mínima dos 100 cartões foi {np.min(dfCartoesAleatorios["rendimento"]):2.2%}')
print(f'O melhor cartao ganhou R$ {np.max(dfCartoesAleatorios["ganho"]):2.2f} e gastou R$ {dfCartoesAleatorios[dfCartoesAleatorios["ganho"] == np.max(dfCartoesAleatorios["ganho"])]["gasto"].values[0]:2.2f}')
```

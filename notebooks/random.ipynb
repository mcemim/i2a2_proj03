```python
concursos = list(df['Concurso'].unique())
jogosSelecionados = []
ganhosSelecionados = []
allSelecionados = []
# proximoJogo = [0]
ganho = 0
gasto = 0
ac11 = 0
ac12 = 0
ac13 = 0
ac14 = 0
ac15 = 0
rendimento = 0
rendimento_max = 0
for i in range(100):
    print(i)
    gasto = 0
    ganho = 0
    proximoJogo = list(np.sort(random.sample(range(1,26), 15)))
    # print(proximoJogo)
    for concurso in concursos:        
        idx = df.index[df['Concurso'] == concurso].tolist()[0]
        gasto += 1
        acertos = 0
        for bola in df.loc[idx, "sorteados_seq"]:
            if bola in proximoJogo:
                acertos += 1
        if acertos > 10:
            coluna = f"Valor Rateio {str(acertos)} Números"
            ganho += df.loc[idx, coluna]
            if acertos == 11:
                ac11 += 1
            if acertos == 12:
                ac12 += 1
            if acertos == 13:
                ac13 += 1
            if acertos == 14:
                ac14 += 1
            if acertos == 15:
                ac15 += 1
    rendimento = (ganho/gasto)-1
    rendimento_max = max(rendimento, rendimento_max)
    print(f'{rendimento} {rendimento_max:.2f}.')
    if (rendimento) > 1:
        jogosSelecionados.append(proximoJogo)
        ganhosSelecionados.append(ganho, [ac11, ac12, ac13, ac14, ac15])
        
        # print(f'Testei {i} combinações e encontrei essa que parece boa: {proximoJogo}')
        # print(f'Ganho total: R$ {ganho:.2f}. Gasto total: R$ {gasto:.2f}. Rendimento de {(ganho/gasto)-1:.2%}                     ')
        # print(f'11 acertos: {ac11}')
        # print(f'12 acertos: {ac12}')
        # print(f'13 acertos: {ac13}')
        # print(f'14 acertos: {ac14}')
        # print(f'15 acertos: {ac15}')
        # break
allSelecionados = [jogosSelecionados, ganhosSelecionados]
```
